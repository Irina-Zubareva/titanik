```python
import pandas as pan
import seaborn as sns
import matplotlib.pyplot as mpl
import re
import numpy as np
import stats
from scipy.stats import ttest_1samp
from scipy.stats import ttest_rel
%matplotlib inline
dat = pan.read_csv ('titanik.csv', index_col='PassengerId')
```

```python
#определение пропущенных значений
print('Подсчет нулевых значений:', '\t', dat.isnull().sum())
sns.heatmap(dat.isnull(), cbar = False).set_title('пропущенные значения')
```

```python
#подсчет уникальных значений (опр категориальных переменных)
print('Уникальные значения:', '\t', dat.nunique())
```

```python
#Подстчет статистик
print('Статистика по возрасту:')
print(dat['Age'].describe())
```

```python
#Задание 2
Fem = Mal = 0
for el in dat['Sex']:
    if el == 'male':
        Mal += 1
    else:
        Fem += 1
print ('Мужчины:', Mal,'Женщины:', Fem)

#Задание 2
V = P = 0
for el in dat['Survived']:
    P += 1
    if el == 1:
        V += 1
Pr = round(100*V/P,2)
print ('Процент выживших:', Pr)

#Задание 3
Perviy = 0
for el in dat['Pclass']:
    if el == 1:
        Perviy += 1
Perv = round(100*Perviy/P,2)
print ('Процент пассажиров первого класса:', Perv)
```

```python
#Задание 4
print('Средний возраст:',round(dat['Age'].mean(axis = 0, skipna = True), 2))
print('Медиана возраста:',dat['Age'].median(axis = 0, skipna = True))

#Задание 6
def first_name(name): #функция для поиска основного имени
    t = re.search('^[^,]+, (.*)', name) #имя пассажира после запятой
    if t:
        name = t.group(1)
    t = re.search('\(([^)]+)\)', name) #женские имена в скобках (если есть скобки)
    if t:
         name = t.group(1)
    name = re.sub('(Miss\. | Mrs\. | Ms\.)', '', name) #заменяем обращения на кавычки
    name = name.split(' ')[0].replace('"',' ') #делим строку с именем и удаляем кавычки (берем самое левое имя)
    return name


names = dat[dat['Sex'] == 'female']['Name'].map(first_name)
names_sch = names.value_counts()
#print('Женские имена и сколько раз они встречаются:')
#print(names_sch)
print('Самое популярное женское имя:',names_sch.head(1).index.values[0])
```

```python
#Новые графики с комментариями по анализу

#Зависимость выживаемости от возраста: среди не выживших преобладают люди 20-25 лет,
#среди выживших - примерно 30 лет, также заметен пик выживаемости детей младшего возраста
sns.catplot(x="Survived", y="Age", kind="violin", data=dat)
mpl.figure()
#Распределение по возрасту показывает, что на корабле было больше всего людей 20-30 лет
#дети были, в основном, младшего возраста, имеются единичные выбросы в самом старшем возрасте
mpl.hist(data = dat, x = 'Age', bins = 40)
mpl.figure() 
#"Ящик с усами" также отражает ранее полученное распределение возраста
sns.boxplot(data = dat, x = 'Age')
mpl.figure()
#Распределение по цене билета показывает, что больше всего было пассажиров, купивших более дешевые билеты
mpl.hist(data = dat, x = 'Fare', bins = 20)
mpl.figure()
#"Ящик с усами" для распределения по цене билета
sns.boxplot(data = dat, x = 'Fare')
mpl.figure()
#"Ящики с усами" для распределения пассажиров по классу в зависимости от возраста с учетом пола.
#В первом классе было больше всего пассажиров и их средний возраст был выше,
#разница в возрасте у мужчин и женщин этого класса была самой большой, чем у всех остальных класов.
#Меньше всего людей выбрали второй класс, они были среднего возраста. Третий класс выбирали молодые люди.
#Можно сделать вывод, что путешествовали преимущественно молодые люди небольшого достатка или достаточно богатые люди в возрасте.
sns.boxplot(data = dat, x = 'Pclass', y = 'Age', hue = 'Sex')
mpl.figure()
```

```python
#Категории
#Категории по цене билет
dat['Fare_Category'] = pan.cut(dat['Fare'], bins=[0,7.90,14.45,31.28,120], labels=['Low','Mid', 'High_Mid','High'])
#Категории по возрасту
dat['Age_Category'] = pan.cut(dat['Age'], bins=[0,10,18,29,35,45,80], labels=['Child','Teen', 'Young','Mid','High_mid', 'Old'])
#Категории по наличию братьев/сестер/супругов на борту
dat['SibSp_Category'] = pan.cut(dat['SibSp'], bins=[-0.5,0.5,5], labels=['Alone_sibsp','NotAlone_sibsp'])
#Категории по одиночеству на борту
dat['Alone_Category'] = pan.cut(dat['Parch']+dat['SibSp'], bins=[-0.5,0.5,10], labels=['travel_alon','travel_notalone'])
print(dat['Alone_Category'])
```

```python
#Новые графики с категориями и коментариями по ананлизу

#Скрипичные диаграммы для анализа выживаемости внутри каждой категории возраста.
#Можно сделать вывод, что внутри категории Child вероятность выжить была больше, чем не выжить;
#Внутри категории Young вероятность не выжить была больше, чем выжить;
#По остальным категориям однозначно точных выводов сделать нельзя, только приближенные. 
sns.catplot(x="Age_Category", y="Survived", kind="violin", data=dat)
mpl.figure()
#Зависимость выживаемости от возрастной категории:
#Распредедление от большего числа выжить к меньшем (а соответственно и от большей вероятности выжить к меньшей):
#Child - Mid - Teen - High_mid - Old - Young
sns.catplot(x="Age_Category", y="Survived", kind="bar", data=dat)
mpl.figure()
#Зависимость цены билета от вероятности выживания: Чем дороже билет, тем вероятнее выживание
sns.catplot(x="Fare_Category", y="Survived", kind="bar", data=dat)
mpl.figure()
#Распределение возраста внутри категории
sns.catplot(x="Age_Category", y="Age", kind="violin", data=dat)
mpl.figure()
#Скрипичная диаграмма выживаемости от наличия братьев/сестер/супругов на борту;
#Если у путешественника не было никого из братьев/сестер/супругов на борту, то он вероятнее погибал, чем выживал
#Сделать вывод по категории, где у путешественников были эти родственники, однозначно нельзя
sns.catplot(x="SibSp_Category", y="Survived", kind="violin", data=dat)
mpl.figure()
#По диаграмме типа catplot можно установить, что вероятнее выжить было, находясь на борту с братьями/сестрами/супругами
sns.catplot(x="SibSp_Category", y="Survived", kind="bar", data=dat)
mpl.figure()
#Скрипичная диаграмма выживаемости от наличия родственников любого типа (братья/сестры/супруги/родители/дети) на борту;
#Если у путешественника не было никого из родственников на борту, то он вероятнее погибал, чем выживал
#Сделать вывод по категории, где у путешественников были родственники, по этой диаграмме однозначно нельзя
sns.catplot(x="Alone_Category", y="Survived", kind="violin", data=dat)
mpl.figure()
#По диаграмме типа catplot можно установить, что вероятнее выжить было, находясь на борту не в одиночку 
sns.catplot(x="Alone_Category", y="Survived", kind="bar", data=dat)
mpl.figure()
#Гистограмма, показывающая зависимость выживаемости от порта посадки;
#Чаще выживали пассажиры, севшие в порту C.
#У пассажиров, севших в порту S была низкая вероятность выжить
sns.catplot(x="Embarked", hue ="Survived", kind="count", data=dat)
mpl.figure()
#Гистограмма, показывающая распределение по ценам билета и порту посадки с учетом выживаемости;
#Чаще выживали пассажиры, имевшие более дорогой билет, вне зависимости от порта посадки
sns.catplot(x="Embarked", y = "Fare", hue ="Survived", kind="bar", data=dat)
mpl.figure()
#Гистограмма, показывающая, билеты какой ценовой категории были у пассажиров того или иного порта;
#В порту C садились пассажиры с билетами преимущественно выше средней цены, 
#в порту S - средней ценовой категории, а в порту Q - низкой ценовой категории
#На основании того, что чаще всего выживали пассажиры порта C, можем сделать вывод, что пассажиры,
#с билетами более дорогих ценовых категории выживали чаще
sns.catplot(x="Embarked", hue ="Fare_Category", kind="count", data=dat)
mpl.figure()
```

```python
#isnull - функция регистрирует пропущенные значения
#transform - функция поддерживает длину результирующего датафрейма равной длине исходного при применении фунции к элементам df
#dropna - функция удаляет пропущенные значения
dat[['Age_1']]=dat[['Age']].fillna(dat.groupby('Sex').transform('mean')) #заполнение пропущенных значений по возрасту
#пустые ячейки заполняются средним значением по полу

dat['Embarked_1']=dat['Embarked'].fillna('C') #пропущенные значения по порту заплняются 'C'
```

```python
sns.pairplot(dat, hue='Sex', palette='Set1');
```

```python
ages = dat['Age_1']
print(ages)
ages_mean = np.mean(ages)
print(ages_mean)
tset, pval = ttest_1samp(ages, 29.7)
print('p-values', pval)
if pval < 0.05: # alpha value is 0.05 or 5%
    print("мы отвергаем нулевую гипотезу")
else:
    print("мы принимаем нулевую гипотезу")
```
